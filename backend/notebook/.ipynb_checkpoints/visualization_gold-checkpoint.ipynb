
# Visualization_gold

```python
# 1Ô∏è‚É£ Import des biblioth√®ques
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style("whitegrid")
```

```python
# 2Ô∏è‚É£ Charger le Gold quotidien
GOLD_PATH = "../gold/gold_daily_risk.csv"
gold_df = pd.read_csv(GOLD_PATH)
gold_df["date"] = pd.to_datetime(gold_df["date"])

print("‚úÖ Donn√©es charg√©es :")
display(gold_df.head())

# 3Ô∏è‚É£ Statistiques rapides
print("\nüìä Statistiques flood risk :")
print(gold_df["flood_risk"].describe())

print("\nüìä Statistiques drought risk :")
print(gold_df["drought_risk"].describe())

# 4Ô∏è‚É£ Histogramme des risques
plt.figure(figsize=(12,5))
sns.histplot(gold_df["flood_risk"], bins=10, color="blue", kde=True)
plt.title("Distribution Flood Risk")
plt.xlabel("Flood Risk")
plt.ylabel("Nombre de jours/zones")
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(gold_df["drought_risk"], bins=10, color="orange", kde=True)
plt.title("Distribution Drought Risk")
plt.xlabel("Drought Risk")
plt.ylabel("Nombre de jours/zones")
plt.show()

# 5Ô∏è‚É£ Visualisation par zone (top 10 flood risk moyens)
top_flood = gold_df.groupby("zone_id")["flood_risk"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_flood.index, y=top_flood.values, palette="Blues_r")
plt.xticks(rotation=45)
plt.title("Top 10 zones avec le plus haut Flood Risk moyen")
plt.ylabel("Flood Risk moyen")
plt.show()

# 6Ô∏è‚É£ Visualisation par zone (top 10 drought risk moyens)
top_drought = gold_df.groupby("zone_id")["drought_risk"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_drought.index, y=top_drought.values, palette="Oranges_r")
plt.xticks(rotation=45)
plt.title("Top 10 zones avec le plus haut Drought Risk moyen")
plt.ylabel("Drought Risk moyen")
plt.show()

# 7Ô∏è‚É£ Carte simplifi√©e (scatter plot latitude/longitude)
plt.figure(figsize=(14,7))
sc = plt.scatter(
    gold_df["longitude"], gold_df["latitude"],
    c=gold_df["flood_risk"], cmap="Blues", s=50, alpha=0.7
)
plt.colorbar(sc, label="Flood Risk")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Flood Risk par position g√©ographique")
plt.show()

plt.figure(figsize=(14,7))
sc = plt.scatter(
    gold_df["longitude"], gold_df["latitude"],
    c=gold_df["drought_risk"], cmap="Oranges", s=50, alpha=0.7
)
plt.colorbar(sc, label="Drought Risk")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Drought Risk par position g√©ographique")
plt.show()
```
